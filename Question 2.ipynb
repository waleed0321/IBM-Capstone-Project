```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plot
import numpy as np
import seaborn as sns
```

```python
df=pd.read_csv('final_project')
df.head()
```

```python
df.info()
```

```python
df['borough'].value_counts()
```

```python
top_complaint = df[df['complaint_type'] == 'HEAT/HOT WATER'].reset_index(drop = True)
top_complaint.head()
```

```python
top=top_complaint['borough'].value_counts()
top.sort_values(ascending=True, inplace=True)
```

```python
top.plot(kind='barh',figsize=(12,12))
```

```python
# finding streets about top complaints
top_complaint_s=top_complaint['street_name'].value_counts()
top_complaint_s.sort_values(ascending=True, inplace=True)
```

```python
top_complaint_s.plot(kind='barh',figsize=(12,12))
```
